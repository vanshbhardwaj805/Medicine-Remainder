```python
import time
from datetime import datetime
from plyer import notification
from playsound import playsound

# 📋 Step 1: Enter your medicine schedule
medicine_schedule = [
    {"name": "Blood Pressure Tablet", "time": "08:00"},
    {"name": "Vitamin C", "time": "12:30"},
    {"name": "Diabetes Tablet", "time": "20:00"}
]

# 🔁 Step 2: Keep checking the time
print("💊 Medicine Reminder Started...")
alerted = set()

while True:
    now = datetime.now().strftime("%H:%M")
    
    for med in medicine_schedule:
        if med["time"] == now and med["time"] not in alerted:
            # 🔔 Notification
            notification.notify(
                title="⏰ Medicine Reminder",
                message=f"It's time to take: {med['name']}",
                timeout=10
            )
            # 🔊 Sound alert
            try:
                playsound("alert.mp3")  # Replace with your alert file
            except:
                print("🔉 Alert sound not found!")

            alerted.add(med["time"])

    time.sleep(30)  # Check every 30 seconds
```
